```python
data["sentence"][0]
```

```python
def data_cleaning(df):
    for i in range(len(df)):
        df["sentence"][i]=df["sentence"][i].lower()
        df["sentence"][i]=re.sub("\[.*\]","",df["sentence"][i])
        df["sentence"][i]=re.sub("\(.*\)","",df["sentence"][i])
    return df
```

```python
data=data_cleaning(data)
```
